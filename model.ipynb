```python
def print_unique_column_values(data):
    for column in data:
        if data[column].dtype=='object':
            print(f'{column}: {data[column].unique()}')
```

```python
print_unique_column_values(data)
```

```python
data.replace('No phone service', 'No', inplace=True)
data.replace('No internet service', 'No', inplace=True)
```

```python
print_unique_column_values(data)
```

```python
columns_with_yes_no=['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']
```

```python
for col in columns_with_yes_no:
    data[col].replace({'Yes':1, 'No':0}, inplace=True)
```

```python
print_unique_column_values(data)
```

```python
for col in data:
    print(f'{col}: {data[col].unique()}')
```

```python
data['gender'].replace({'Male':1, 'Female':0}, inplace=True)
data['gender'].unique()
```

```python
new_data=pd.get_dummies(data, columns=['InternetService', 'Contract', 'PaymentMethod'])
```

```python
new_data.head()
```

```python
new_data.dtypes
```

```python
cols_to_scale=['tenure', 'MonthlyCharges', 'TotalCharges']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
new_data[cols_to_scale]=scaler.fit_transform(new_data[cols_to_scale])
```

```python
new_data.sample(5)
```

```python
X=new_data.drop('Churn',axis='columns')
X.head()
```

```python
y=new_data['Churn']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras
model=keras.Sequential([
    keras.layers.Dense(20, input_shape=(26,), activation='relu'),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp=model.predict(X_test)
yp[:5]
```

```python
y_test[:10]
```

```python
y_pred=[]
for element in yp:
    if element>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:20]
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test, y_pred)
```

```python
print(cr)
```

```python
cr=tf.math.confusion_matrix(y_test, y_pred)
print(cr)
```

```python
import seaborn as sns
sns.heatmap(cr, annot=True, fmt='d')
plt.xlabel('Predictions')
plt.ylabel('Truth')
```

```python
import numpy as np
```

```python
def ANN(X_train, y_train, X_test, y_test, loss, weights):
    model=keras.Sequential([
        keras.layers.Dense(20, input_shape=(26,), activation='relu'),
        keras.layers.Dense(20, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])
    if weights==-1:
          model.fit(X_train, y_train, epochs=100)
    else:
        model.fit(X_train, y_train, epochs=100, class_weight=weights)
    
    print(model.evaluate(X_test, y_test))
    
    y_pred=np.round(model.predict(X_test))
    print(classification_report(y_test, y_pred))
    
```

```python
ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```
